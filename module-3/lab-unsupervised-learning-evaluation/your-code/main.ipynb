Complete the exercises below to solidify your knowledge and understanding of unsupervised learning model evaluation.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
from sklearn import cluster
from yellowbrick.cluster import KElbowVisualizer
```

```python
from sklearn import datasets

data = datasets.load_wine()

X = pd.DataFrame(data["data"], columns=data["feature_names"])
y = pd.Series(data["target"])
```

## 1. Train a KMeans clustering model on the data set using 8 clusters and compute the silhouette score for the model.

```python
kmeans = cluster.KMeans(n_clusters=8)
kmeans.fit(X)
y_pred = kmeans.predict(X)

print("Model KMeans Silhouette Score: {}".format(silhouette_score(X, y_pred)))
```

## 2. Train a KMeans clustering model on the data set using 5 clusters and compute the silhouette score for the model.

```python
kmeans = cluster.KMeans(n_clusters=5)
kmeans.fit(X)
y_pred = kmeans.predict(X)

print("Model KMeans Silhouette Score: {}".format(silhouette_score(X, y_pred)))
```

## 3. Train a KMeans clustering model on the data set using 3 clusters and compute the silhouette score for the model.

```python
kmeans = cluster.KMeans(n_clusters=3)
kmeans.fit(X)
y_pred = kmeans.predict(X)

print("Model KMeans Silhouette Score: {}".format(silhouette_score(X, y_pred)))
```

## 4. Use elbow curve visualizations to see if you can determine the best number of clusters to use.

The Yellowbrick library has 3 metrics that you can plot using the `metric` parameter:

- **distortion**: mean sum of squared distances to centers
- **silhouette**: mean ratio of intra-cluster and nearest-cluster distance
- **calinski_harabaz**: ratio of within to between cluster dispersion

```python
from sklearn.cluster import KMeans

model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,12), metric='distortion')
visualizer.fit(X)
visualizer.poof()
```

```python
model = KMeans()
visualizer = KElbowVisualizer(model, k=(2,12), metric='silhouette')
visualizer.fit(X)
visualizer.poof()
```

```python
model = KMeans()
visualizer = KElbowVisualizer(model, k=(2,12), metric='calinski_harabaz')
visualizer.fit(X)
visualizer.poof()
```

## 5. Try performing the same elbow tests with an AgglomerativeClustering model and compare the results you get to the KMeans results.

```python
single = cluster.AgglomerativeClustering(n_clusters=3, linkage='single')
single.fit(X)
y_pred = single.labels_.astype(np.int)
print("Model AgglomerativeClustering Silhouette Score: {}".format(silhouette_score(X, y_pred)))
```

```python
from sklearn.cluster import AgglomerativeClustering

model = AgglomerativeClustering()
visualizer = KElbowVisualizer(model, k=(2,12), metric='silhouette')
visualizer.fit(X)
visualizer.poof()
```

## 6. Create and plot a scatter matrix showing how the clusters are grouped across all the different combinations of variables in the data.

Use the model and number of clusters that returned the best result above.

```python
kmeans = cluster.KMeans(n_clusters=3)
data_clusters = kmeans.fit(X)
X['labels'] = data_clusters.labels_
```

```python
import seaborn as sns
sns.set(style="ticks")

sns.pairplot(X, hue='labels')
```

## 7. Apply a PCA transform and plot the first two principle components with the plot point colors determined by cluster.

```python
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(X)
pca_X = pca.transform(X)

fig = plt.figure(figsize=(6, 6))
ax = [plt.subplot2grid((6, 1), (0, 0), rowspan=4), plt.subplot2grid((6, 1), (5, 0))]

ax[0].scatter(pca_X[:, 0], pca_X[:, 1], c=X.labels)
ax[0].set_xlabel("PCA 1")
ax[0].set_ylabel("PCA 2")

ax[1].set_title('PCA 3', pad=-5)
ax[1].hlines(1, -7, 10)
y = np.ones(X.shape[0])
ax[1].plot(pca_X[:, 2], y, '|', ms='20')
ax[1].axis('off')

plt.show()
```

```python
fig = plt.figure(figsize=(6, 6))
ax = [plt.subplot2grid((6, 1), (0, 0), rowspan=4), plt.subplot2grid((6, 1), (5, 0))]

ax[0].scatter(pca_X[:, 0], pca_X[:, 2], c=X.labels)
ax[0].set_xlabel("PCA 1")
ax[0].set_ylabel("PCA 3")

ax[1].set_title('PCA 2', pad=-5)
ax[1].hlines(1, -7, 10)
y = np.ones(X.shape[0])
ax[1].plot(pca_X[:, 1], y, '|', ms='20')
ax[1].axis('off')

plt.show()
```

## 8. Generate a series of t-SNE plots showing the clusters at a variety of perplexities.

```python
from sklearn.manifold import TSNE

fig, ax = plt.subplots(1, 4, figsize=(15, 7))
for i, perp in enumerate([5, 30, 50, 100]):
    tsne = TSNE(perplexity=perp)
    x_embedded = tsne.fit_transform(X)
    ax[i].scatter(x_embedded[:, 0], x_embedded[:, 1], c=X.labels)
    ax[i].set_title("Perplexity = {}".format(perp))
    
plt.show()
```
