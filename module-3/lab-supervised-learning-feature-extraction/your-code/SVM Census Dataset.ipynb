```python
import pandas as pd
census = pd.read_csv("~/Desktop/census.csv")
census.head()
```

```python
doctors = pd.get_dummies(census.doctor_name, drop_first=True)
census_dummy = pd.concat([census, doctors], axis=1)
census_dummy.head()
```

```python
df = census_dummy._get_numeric_data()
df.drop(['patient_nhs_number','id'],axis=1,inplace=True)
df.head()
print(df.columns)
df['diabetic'] = pd.get_dummies(df.patient_diabetic,drop_first=True)
df['allergic'] = pd.get_dummies(df.patient_allergic,drop_first=True)
df['regular_visit'] = pd.get_dummies(df.is_regular_visit,drop_first=True)
df.drop(['patient_diabetic','patient_allergic','patient_show','is_regular_visit'],axis=1,inplace=True)
df.head()
```

```python
X_train = df[0:800]
X_test = df[800:1000]
print(X_train.shape,X_test.shape)
```

```python
y = df.allergic
X = df.drop(['allergic'], axis=1)
display(y.head())
display(X.head())
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape )
```

```python
from sklearn import svm
clf = svm.SVC()
clf.fit(X_train, y_train)
```

```python
np.sum(np.abs(y_test.astype('int64') - clf.predict(X_test).astype('int64')))
```
