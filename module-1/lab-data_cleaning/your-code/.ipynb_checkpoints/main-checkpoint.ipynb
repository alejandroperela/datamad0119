#### 4. Import the users table 

```python
users = pd.read_sql_query('SELECT * FROM stats.users', engine)
```

#### 5. Rename Id column to userId

```python
users = users.rename(index=str, columns={"Id": "userId"})
users.head()
```

#### 6. Import the posts table. 

```python
posts = pd.read_sql_query('SELECT * FROM stats.posts', engine)
```

#### 7. Rename Id column to postId and OwnerUserId to userId

```python
posts = posts.rename(index=str, columns={"Id": "postId", "OwnerUserId": "userId"})
posts.head()
```

#### 8. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userID,ViewCount,CommentCount

```python
users_new = users[['userId', 'Reputation', 'Views', 'UpVotes', 'DownVotes']].copy()
posts_new = posts[['postId', 'Score', 'userId', 'ViewCount', 'CommentCount']].copy()
users_new.head()
posts_new.head()
```

#### 8. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
users_posts = pd.merge(users_new, posts_new, on='userId')
users_posts.head(10)
```

#### 9. How many missing values do you have in your merged dataframe? On which columns?

```python
null_cols = users_posts.isnull().sum()
null_cols[null_cols > 0]
```

#### 10. You will need to make something with missing values.  Will you clean or filling them? Explain. 
**Remember** to check the results of your code before passing to the next step

```python
users_posts[['ViewCount']] = users_posts[['ViewCount']].fillna(0)
print(set(users_posts['ViewCount']))
```

#### 11. Adjust the data types in order to avoid future issues. Which ones should be changed? 

```python
users_posts.dtypes
users_posts['ViewCount'] = users_posts['ViewCount'].astype('int64')
users_posts['ViewCount'].dtype
```

#### Bonus: Identify extreme values in your merged dataframe as you have learned in class, create a dataframe called outliers with the same columns as our data set and calculate the bounds. The values of the outliers dataframe will be the values of the merged_df that fall outside that bounds. You will need to save your outliers dataframe to a csv file on your-code folder.

```python
stats = users_posts.describe().transpose()
stats['IQR'] = stats['75%'] - stats['25%']

outliers = pd.DataFrame(columns=users_posts.columns)

for col in stats.index:
    iqr = stats.at[col,'IQR']
    cutoff = iqr * 1.5
    lower = stats.at[col,'25%'] - cutoff
    upper = stats.at[col,'75%'] + cutoff
    results = users_posts[(users_posts[col] < lower) | 
                   (users_posts[col] > upper)].copy()
    results['Outlier'] = col
    outliers = outliers.append(results)
outliers.head()

outliers = outliers.to_csv('/home/alejandro/IRONHACK/LABORATORIOS/datamad0119/module-1/lab-data_cleaning/your-code/outliers.csv', index=False)
```
