#### We suspect that there is a linear relationship between `Fare` and `Age`. Use the appropriate plot to show the relationship between these 2 features. There are 2 ways, please do it both ways.

*Hint: One of the ways involves using Seaborn*

```python
#your code-1st way

import plotly.plotly as py
import plotly.graph_objs as go

# MatPlotlib
import matplotlib.pyplot as plt
from matplotlib import pylab

# Scientific libraries
from numpy import arange,array,ones
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(titanic['Fare'],titanic['Age'])
line = slope*titanic['Fare']+intercept

plt.plot(titanic['Fare'],titanic['Age'],'o', titanic['Fare'], line)
pylab.title('Linear Fit with Matplotlib')
ax = plt.gca()
ax.set_axis_bgcolor((0.898, 0.898, 0.898))
fig = plt.gcf()
py.plot_mpl(fig, filename='linear-Fit-with-matplotlib')
```

```python
# your code-2nd way
sns.regplot(x="Fare", y="Age", data=titanic)
```
